```python
import numpy as np
import pandas as pd
# sigmoid function to normalize inputs
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# sigmoid derivatives to adjust synaptic weights
def sigmoid_derivative(x):
    return x * (1 - x)
```

```python
# input dataset
dataset = pd.read_csv('ds2_train.csv')
training_inputs = dataset.drop('y', axis=1) 
# output dataset
training_outputs = dataset['y']
training_inputs=(training_inputs.values)
training_outputs=(training_outputs.values)
training_outputs=training_outputs.reshape(800,1)
#raining_outputs=(training_outputs.values)
print(training_inputs.shape)
print(training_outputs.shape)
```

```python
np.random.seed(1)
synaptic_weights = 2 * np.random.random((2,1)) - 1
```

```python
print('Random starting synaptic weights: ')
print(synaptic_weights)
```

```python
for iteration in range(10):

    # Define input layer
    input_layer = training_inputs
    # Normalize the product of the input layer with the synaptic weights
    outputs = sigmoid(np.dot(input_layer, synaptic_weights))
   
    # how much did we miss?
    error = training_outputs - outputs

    # multiply how much we missed by the
    # slope of the sigmoid at the values in outputs
    adjustments = error * sigmoid_derivative(outputs)

    # update weights
    synaptic_weights += np.dot(input_layer.T, adjustments)
```

```python
print('Synaptic weights after training: ')
print(synaptic_weights)

print("Output After Training:")
print(outputs)
```
